```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

### Creating a line chart from a list object

#### Plotting a line chart in matplotlib

```python
import random
x = range(1,10)
y = list(range(1,10))
random.shuffle(y)
plt.plot(x, y)
```

#### Plotting a line chart from a Pandas object

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)
print(cars.shape)
print(cars.columns)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
print(cars.columns)
mpg = cars['mpg']
cars.head()
```

```python
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

### Creating bar charts
#### Creating a bar chart from a list

```python
plt.bar(x, y)
```

#### Creating bar charts from Pandas objects

```python
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

### Creating a pie chart

```python
x = [1,2,3,4,0.5]
plt.pie(x)
```

```python
print(cars['am'].value_counts())
am_type = ['0', '1']
plt.pie(list(cars['am'].value_counts()), labels = am_type)
```

```python
cars['cyl'].value_counts()
```

```python
cyl_type = ['8 cyl', '4 cyl','6 cyl',]

plt.pie(list(cars['cyl'].value_counts()))
plt.legend(cyl_type, loc='best')
```

### Saving a plot

```python
plt.savefig('pie_chart.png')
plt.show()
```
